### World Happiness Index by Reports [2013-2023]:

#### Introduction:

Happiness Index, in brief, is a measure used to assess the overall happiness and life satisfaction of individuals or populations in a specific country. It typically quantifies people's subjective well-being and contentment with life through surveys or questionnaires.

The World Happiness Report includes several other variables that are used to analyze and understand the level of happiness in countries. Some of these variables include:

1. *Social Support:* Measures the level of social support individuals receive from friends, family, and the community.

2. *Gross National Income:* Measures the total income per capita in the country.

3. *Health:* Reflects the overall health status and the availability of healthcare services.

4. *Freedom to Make Life Choices:* Measures the freedom of individuals to make life decisions.

5. *Corruption Rates:* Used to gauge the extent of corruption in government and institutions.

These variables are used to understand the factors influencing the level of happiness in different countries and aid in the analysis and interpretation of the data.

These data cover the period from 2013 to 2022 and represent 137 countries from around the world.

In this example, I will be testing hypotheses between the Happiness Index and Social Support. This analysis aims to explore and quantify the relationship between individuals' self-reported happiness levels (Happiness Index) and the extent of social support they receive. By statistically examining these variables, we seek to ascertain whether there exists a significant association that can offer insights into the factors contributing to overall happiness within the dataset.


#### Hypothesis Testing: Null and Alternative Hypotheses:

 - Null Hypothesis (H0): There is no significant correlation between social support and the level of happiness.
 - Alternative Hypothesis (H1): There is a significant correlation between social support and the level of happiness.


#### Names of the independent and dependent variable (s):

* Independent Variable: "Happiness score".
* Dependent Variable: "Social support".

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
data = files.upload()
```

```python
data = pd.read_excel('WHR2023.xlsx')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data = data[['Country_name','Ladder_score','Logged_GDP_per_capita','Social_support','Healthy_life_expectancy',
             'Freedom_to_make_life_choices', 'Generosity', 'Perceptions_of_corruption']]
data
```

```python
corr = data.corr(method = 'pearson')
corr
```

```python
## Heatmap plot:
plt.figure(figsize = (15,5))
sns.heatmap(corr, cmap ='Spectral', annot = True, fmt=".2f", vmin = -1 , vmax =1)

plt.show()
```

** Ladder score is Happiness Score

 The correlation between social support and the happiness score, which is r =  0.83

```python
plt.title('Scatter Plot between Happiness Score and Healthy life expectancy')
sns.scatterplot(x = data.Ladder_score, y = data.Social_support)
plt.xlabel('Happiness Score')
plt.ylabel('Social support')
```

### Test Statistics: The Pearson correlation coefficient (Pearson's r). It measures the linear correlation between two variables. In this case, the correlation between social support and the happiness score, which is r =  0.83, indicates a strong positive correlation. Therefore, the results accept H1 and reject H0.
